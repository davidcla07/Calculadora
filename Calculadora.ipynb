En el siguiente código se crean las funciones de operaciones básicas:

```python
def suma():
    print("\nEn este módulo puede sumar cualquier cantidad de valores que necesite\n")
    cont = 0
    while True:
        opcion = input("\nIngrese 1 para continuar, 2 para regresar al menú\n")
        if opcion == "1":
            if cont == 0:
                valor = input("\nIngrese el primer valor\n")
                try:
                    resultado = float(valor)
                except:
                    print("\Ingrese un valor válido\n")
            if cont > 0:
                valor = input("\nIngrese un valor para sumar\n")
                try:
                    valor = float(valor)
                    resultado += valor
                    print(f'La suma de los valores es : {resultado:.2f}')
                except:
                    print("\nIngrese un valor válido\n")
        elif opcion == "2":
            break
        else:
            print("\nIngrese un valor válido\n")
        cont += 1


def resta():
    print("\nEn este módulo puede restar cualquier cantidad de valores que necesite\n")
    cont = 0
    while True:
        opcion = input("\nIngrese 1 para continuar, 2 para regresar al menú\n")
        if opcion == "1":
            if cont == 0:
                valor = input("\nIngrese el primer valor\n")
                try:
                    resultado = float(valor)
                except:
                    print("\Ingrese un valor válido\n")
            if cont > 0:
                valor = input("\nIngrese un valor para restar\n")
                try:
                    valor = float(valor)
                    resultado -= valor
                    print(f'La resta de los valores es : {resultado:.2f}')
                except:
                    print("\nIngrese un valor válido\n")
        elif opcion == "2":
            break
        else:
            print("\nIngrese un valor válido\n")
        cont += 1


def multiplicacion():
    print("\nEn este módulo puede multiplicar cualquier cantidad de valores que necesite\n")
    cont = 0
    while True:
        opcion = input("\nIngrese 1 para continuar, 2 para regresar al menú\n")
        if opcion == "1":
            if cont == 0:
                valor = input("\nIngrese el primer valor\n")
                try:
                    resultado = float(valor)
                except:
                    print("\Ingrese un valor válido\n")
            if cont > 0:
                valor = input("\nIngrese un valor para multiplicar\n")
                try:
                    valor = float(valor)
                    resultado *= valor
                    print(f'La multiplicación de los valores es : {resultado:.2f}')
                except:
                    print("\nIngrese un valor válido\n")
        elif opcion == "2":
            break
        else:
            print("\nIngrese un valor válido\n")
        cont += 1

    
def division():
    print("\nEn este módulo puede dividir cualquier cantidad de valores que necesite\n")
    cont = 0
    while True:
        opcion = input("\nIngrese 1 para continuar, 2 para regresar al menú\n")
        if opcion == "1":
            if cont == 0:
                valor = input("\nIngrese el primer valor\n")
                try:
                    resultado = float(valor)
                except:
                    print("\Ingrese un valor válido\n")
            if cont > 0:
                valor = input("\nIngrese un valor para dividir\n")
                try:
                    valor = float(valor)
                    resultado /= valor
                    print(f'La división de los valores es : {resultado:.2f}')
                except:
                    print("\nIngrese un valor válido\n")
        elif opcion == "2":
            break
        else:
            print("\nIngrese un valor válido\n")
        cont += 1
```

Ahora creamos las funciones para las operaciones aritmeticas extendidas:

```python
 import math

def division_entera():
    print("\nEn este módulo puede calcular la division entera de los valores que necesite\n")
    while True:       
        opcion = input("\nIngrese 1 para continuar, 2 para regresar al menú\n")
        if opcion == "1":
            dividendo = input("\nIngrese el numero a dividir (dividendo)\n")
            divisor = input("\nIngrese el numero con el cual divide (divisor)\n")
            try:
                resultado = float(dividendo)//float(divisor)
                print(f'\nLa division entera entre {dividendo} y {divisor} es : {resultado}')
            except:
                print("\nIngrese un valor válido\n")
        elif opcion == "2":
            break
        else:
            print("\nIngrese un valor válido\n")

def residuo():
    print("\nEn este módulo puede calcular el residuo de la division de los valores que necesite\n")
    while True:       
        opcion = input("\nIngrese 1 para continuar, 2 para regresar al menú\n")
        if opcion == "1":
            dividendo = input("\nIngrese el numero a dividir (dividendo)\n")
            divisor = input("\nIngrese el numero con el cual divide (divisor)\n")
            try:
                resultado = float(dividendo)%float(divisor)
                print(f'\nEl residuo entre {dividendo} y {divisor} es : {resultado}')
            except:
                print("\nIngrese un valor válido\n")
        elif opcion == "2":
            break
        else:
            print("\nIngrese un valor válido\n")


def exponenciacion():
    print("\nEn este módulo puede calcular el exponente de los valores que necesite\n")
    while True:       
        opcion = input("\nIngrese 1 para continuar, 2 para regresar al menú\n")
        if opcion == "1":
            valor = input("\nIngrese el valor a calcular\n")
            try:
                resultado = math.exp(float(valor))
                print(f'\nEl exponencial de {valor} es : {resultado:.2f}')
            except:
                print("\nIngrese un valor válido\n")
        elif opcion == "2":
            break
        else:
            print("\nIngrese un valor válido\n")

            
def raiz():
    print("\nEn este módulo puede calcular la raiz de los valores que necesite\n")
    while True:       
        opcion = input("\nIngrese 1 para continuar, 2 para regresar al menú\n")
        if opcion == "1":
            raiz = input("\nIngrese el numero de la raiz\n")
            valor = input("\nIngrese el valor a calcular la raiz\n")
            try:
                resultado = math.pow(float(valor),1/int(raiz))
                print(f'\nLa raiz a la {raiz} de {valor} es : {resultado:.2f}')
            except:
                print("\nIngrese un valor válido\n")
        elif opcion == "2":
            break
        else:
            print("\nIngrese un valor válido\n")
            

def logaritmo_10():
    print("\nEn este módulo puede calcular el logaritmo base 10 de los valores que necesite\n")
    while True:       
        opcion = input("\nIngrese 1 para continuar, 2 para regresar al menú\n")
        if opcion == "1":
            valor = input("\nIngrese el valor a calcular el logaritmo\n")
            try:
                resultado = math.log10(float(valor))
                print(f'\nEl logaritmo base 10 de {valor} es : {resultado:.2f}')
            except:
                print("\nIngrese un valor válido\n")
        elif opcion == "2":
            break
        else:
            print("\nIngrese un valor válido\n")
            

def logaritmo():
    print("\nEn este módulo puede calcular el logaritmo de los valores que necesite\n")
    while True:       
        opcion = input("\nIngrese 1 para continuar, 2 para regresar al menú\n")
        if opcion == "1":
            valor = input("\nIngrese el valor a calcular el logaritmo\n")
            base = input("\nIngrese el valor de la base para calcular el logaritmo\n")
            try:
                resultado = math.log(float(valor), float(base))
                print(f'\nEl logaritmo de {valor} con base {base} es : {resultado:.2f}')
            except:
                print("\nIngrese un valor válido\n")
        elif opcion == "2":
            break
        else:
            print("\nIngrese un valor válido\n")
            
            
def absoluto():
    print("\nEn este módulo puede calcular el valor absoluto de los valores que necesite\n")
    while True:       
        opcion = input("\nIngrese 1 para continuar, 2 para regresar al menú\n")
        if opcion == "1":
            valor = input("\nIngrese el valor a calcular el valor absoluto\n")
            try:
                resultado = abs(float(valor))
                print(f'\nEl absoluto de {valor} es : {resultado:.2f}')
            except:
                print("\nIngrese un valor válido\n")
        elif opcion == "2":
            break
        else:
            print("\nIngrese un valor válido\n")
            

def inverso():
    print("\nEn este módulo puede calcular el inverso de los valores que necesite\n")
    while True:       
        opcion = input("\nIngrese 1 para continuar, 2 para regresar al menú\n")
        if opcion == "1":
            valor = input("\nIngrese el valor a calcular el inverso\n")
            try:
                resultado = 1/(float(valor))
                print(f'\nEl inverso de {valor} es : {resultado:.2f}')
            except:
                print("\nIngrese un valor válido\n")
        elif opcion == "2":
            break
        else:
            print("\nIngrese un valor válido\n")
            

def factorial():
    print("\nEn este módulo puede calcular el factorial de los valores que necesite\n")
    while True:       
        opcion = input("\nIngrese 1 para continuar, 2 para regresar al menú\n")
        if opcion == "1":
            valor = input("\nIngrese el valor a calcular el valor factorial\n")
            try:
                resultado = math.factorial(int(valor))
                print(f'\nEl factorial de {valor} es : {resultado:.2f}')
            except:
                print("\nIngrese un valor válido\n")
        elif opcion == "2":
            break
        else:
            print("\nIngrese un valor válido\n")
```

A diferencia de las funciones básicas, en la trigonometrica es poco frecuente realizar las operaciones acumulativas, por lo cual en este menú no se establecera la opción de mantener el resultado para operarlo con la misma función.

```python
import math

def seno():
    print("\nEn este módulo puede calcular el Seno de los valores que necesite\n")
    while True:       
        opcion = input("\nIngrese 1 para continuar, 2 para regresar al menú\n")
        if opcion == "1":
            valor = input("\nIngrese el numero de grados para calcular el Seno\n")
            try:
                resultado = math.sin(math.radians(int(valor)))
                print(f'\nEl Seno de {valor} es : {resultado:.2f}')
            except:
                print("\Ingrese un valor válido\n")
        elif opcion == "2":
            break
        else:
            print("\nIngrese un valor válido\n")
        

def coseno():
    print("\nEn este módulo puede calcular el Coseno de los valores que necesite\n")
    while True:       
        opcion = input("\nIngrese 1 para continuar, 2 para regresar al menú\n")
        if opcion == "1":
            valor = input("\nIngrese el numero de grados para calcular el Coseno\n")
            try:
                resultado = math.cos(math.radians(int(valor)))
                print(f'\nEl Coseno de {valor} es : {resultado:.2f}')
            except:
                print("\Ingrese un valor válido\n")
        elif opcion == "2":
            break
        else:
            print("\nIngrese un valor válido\n")

def tangente():
    print("\nEn este módulo puede calcular la tangente de los valores que necesite\n")
    while True:       
        opcion = input("\nIngrese 1 para continuar, 2 para regresar al menú\n")
        if opcion == "1":
            valor = input("\nIngrese el numero de grados para calcular la tangente\n")
            try:
                resultado = math.tab(math.radians(int(valor)))
                print(f'\nLa tangente de {valor} es : {resultado:.2f}')
            except:
                print("\Ingrese un valor válido\n")
        elif opcion == "2":
            break
        else:
            print("\nIngrese un valor válido\n")
   
```

Lo siguiente es crear las funciones de estadisticas. Se resalta que el promedio es una medida aritmetica que se calcula sumando los valores y dividiendo por la cantidad de estos, mientras que la media es una medida de dispersión, que se refiere al valor esperado de una población. El valor esperado se calcula sumando los productos del valor de cada clase por su propabilidad de ocurrencia. Para el cálculo de la mediana se utilizan las formulas de dispersión encontradas en la literatura, dependiendo de la cantidad de valores ingresados.

De esta forma se crean las funciones.

```python
import numpy as np

def promedio():
    print("\nEn este módulo puede calcular el promedio de cualquier cantidad de valores que necesite\n")
    cont = 0
    while True:
        opcion = input("\nIngrese 1 para continuar, 2 para regresar al menú\n")
        if opcion == "1":
            if cont == 0:
                valor = input("\nIngrese el primer valor\n")
                try:
                    resultado = float(valor)
                except:
                    print("\Ingrese un valor válido\n")
            if cont > 0:
                valor = input("\nIngrese un nuevo valor para promediar\n")
                try:
                    valor = float(valor)
                    resultado += valor
                    promedio = resultado/(cont+1)
                    print(f'El promedio de los valores es : {promedio:.2f}')
                except:
                    print("\nIngrese un valor válido\n")
        elif opcion == "2":
            break
        else:
            print("\nIngrese un valor válido\n")
        cont += 1


def media():
    print("\nEn este módulo puede calcular la media de los valores que necesite\n")
    list_values = list()
    while True:       
        opcion = input("\nIngrese 1 para adicionar valores, 2 para obtener el resultado y regresar al menú\n")
        if opcion == "1":
            valor = input("\nIngrese un valor númerico\n")
            try:
                valor = float(valor)
                list_values.append(valor)
            except:
                print("\Ingrese un valor válido\n")
        elif opcion == "2":
            values_unique = np.unique(list_values).tolist()
            dictionary = {}
            
            for value in values_unique:
                count_value = list_values.count(value)
                dictionary[value] = count_value
                
                expected_value= 0
                for value, count in dictionary.items():
                    prob_value = count/len(list_values)
                    producto = value * prob_value 
                    expected_value += producto
            print(f"La media de los valores es : {expected_value:.2f}")    
            break
        else:
            print("\nIngrese un valor válido\n")
                


def mediana():
    print("\nEn este módulo puede calcular la mediana de los valores que necesite\n")
    list_values = list()
    while True:       
        opcion = input("\nIngrese 1 para adicionar valores, 2 para obtener el resultado y regresar al menú\n")
        if opcion == "1":
            valor = input("\nIngrese un valor númerico\n")
            try:
                valor = float(valor)
                list_values.append(valor)
            except:
                print("\Ingrese un valor válido\n")
        elif opcion == "2":
            list_values.sort()
            if len(list_values) % 2 != 0:
                mediana = int((len(list_values)+1)/2 - 1)
            else:
                a = int(len(list_values)/2 - 1)
                b = int(len(list_values)/2)
                mediana = (list_values[a]+list_values[b])/2
                
            print(f"La mediana de los valores es : {mediana:.2f}")    
            break
        else:
            print("\nIngrese un valor válido\n")
            

def moda():
    print("\nEn este módulo puede calcular la moda de los valores que necesite\n")
    list_values = list()
    while True:       
        opcion = input("\nIngrese 1 para adicionar valores, 2 para obtener el resultado y regresar al menú\n")
        if opcion == "1":
            valor = input("\nIngrese un valor númerico\n")
            try:
                valor = int(valor)
                list_values.append(valor)
            except:
                print("\Ingrese un valor válido\n")
        elif opcion == "2":
            result = max(set(list_values), key=list_values.count)                
            print(f"La moda de los valores es : {result:.2f}")    
            break
        else:
            print("\nIngrese un valor válido\n")

```

Al igual que con las operaciones basicas, creamos los sub menus para para las otros tipos de operaciones creadas:

Por último creamos el menú principal que funcionará como la calculadora:

```python
def calculadora():
    while True:
        print("\n\nIngrese número de opción para realizar operaciones:\n1.Operaciones básicas\n"\
          "2.Operaciones extendidas\n3.Operaciones trigonometricas\n4.Estadisticas\n5.Salir")
        opcion = input()    
        if opcion == "1":
            operaciones_basicas()
        elif opcion == "2":
            operaciones_extendidas()
        elif opcion == "3":
            trigonometricas()
        elif opcion == "4":
            estadisticas()
        elif opcion == "5":
            break
        else:
            print("Ingrese una opción válida")
```
